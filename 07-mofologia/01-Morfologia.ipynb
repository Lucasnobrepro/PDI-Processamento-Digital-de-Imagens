```python
#Bibliotecas
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
#lendo imagem
im = cv2.imread('flag_binary.jpg')
```

```python
#tipo do elemento estrutural
elem_type = cv2.MORPH_RECT

#Elemento estruturante
element = cv2.getStructuringElement(elem_type,(3,3))

# Criando um elemento estruturante
element1 = np.uint8(np.array([[ 0,  255,  0 ],
                              [255, 255, 255],
                              [ 0,  255,  0 ]]))

#Eroso
im_erode = cv2.erode(im,element)

#Dilatao
im_dilate = cv2.dilate(im, element)

# Bordas
borda1 = cv2.subtract(im, im_erode)
borda2 = cv2.subtract(im_dilate,im)

# tentativa
saida = cv2.dilate(borda1,element1)

# Configurando tamanho do plot
plt.figure(figsize=(50,50))
# plot
plt.subplot(321), plt.imshow(borda1);
plt.title('im - im_erode')
plt.subplot(322), plt.imshow(borda2);
plt.title('im_dilate - im')
plt.subplot(323), plt.imshow(im_dilate);
plt.title('im_dilate')
plt.subplot(324), plt.imshow(im_erode);
plt.title('im_erode')
plt.subplot(325), plt.imshow(im);
plt.title('Imagem original');

# Salva a imagem no diretorio do arquivo atual.
# cv2.imwrite('saida.jpg',saida)
# cv2.imwrite('borda1.jpg',borda1)
```
